```python

import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
DATA_PATH = "vision_fig1.csv"

df = pd.read_csv(DATA_PATH)
```

```python
df['Strategy'] = df['Teacher'].apply(lambda x: "Single" if "Multi" in x else 'Multi')
```

```python

fig = plt.figure(constrained_layout=True)

subfigs = fig.subfigures(2, 1)
```

```python

sns.set(style='whitegrid')

colors = ["#090364", "#4374B3"]
sns.set_palette(sns.color_palette(colors))
df = df.sort_values('Task')
g = sns.catplot(x='Teacher', y='ACC', data=df, col='Task', kind='bar', sharey=False, hue='Strategy', legend=False, height=8, aspect=0.6)

g.set_titles(template='{col_name}', fontweight='bold')
# title template

axes = g.axes.flatten()

for k, (name, group) in enumerate(df.groupby('Task')):
    mmin, mmax = group['ACC'].min(), group['ACC'].max()
    
    axes[k].set_ylim(mmin - 0.05 *mmin, mmax + 0.01 * mmax)
    # rotate x labels
    for item in axes[k].get_xticklabels():
        item.set_rotation(90)
        
    # make x labels bigger
    for item in axes[k].get_xticklabels():
        item.set_fontsize(25)
        item.set_fontweight('bold')
        
    # make y labels bigger
    for item in axes[k].get_yticklabels():
        item.set_fontsize(18)
        
    # remove x label
    axes[k].set_xlabel('')
    # remove y label
    axes[k].set_ylabel('')
        
        
    
    # make title bigger
    axes[k].set_title(axes[k].get_title(), fontsize=20)




# export save
g.fig.tight_layout()

plt.savefig("img_single_teacher.png", dpi=300)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a 2x1 figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 16))

# First catplot on ax1
sns.catplot(x='Teacher', y='ACC', data=df, col='Task', kind='bar', sharey=False, hue='Strategy', legend=False, height=8, aspect=0.6, col_wrap=4)

ax1.set_title('First Catplot')

# Second catplot on ax2
sns.catplot(x='Teacher', y='ACC', data=df, col='Task', kind='bar', sharey=False, hue='Strategy', legend=False, height=8, aspect=0.6, col_wrap=4)

ax2.set_title('Second Catplot')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure and a grid of subplots with specified sizes
fig, axs = plt.subplots(2, 4, figsize=(12, 8))

# First row: 4 columns
for i in range(4):
    axs[0, i].plot([0, 1, 2], [0, i, 2*i], label=f'Row 1, Col {i+1}')
    axs[0, i].set_title(f'Plot 1-{i+1}')
    axs[0, i].legend()

# Second row: 3 columns, leaving the last subplot blank
for i in range(3):
    axs[1, i].plot([0, 1, 2], [i, 1, i**2], label=f'Row 2, Col {i+1}')
    axs[1, i].set_title(f'Plot 2-{i+1}')
    axs[1, i].legend()

# Remove the last (empty) subplot in the second row
fig.delaxes(axs[1, 3])

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```
